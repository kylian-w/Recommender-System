```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
from sklearn import preprocessing
```

```python
df=pd.read_csv('filtered_cb_df.csv')
df['new'] = df.index # I add a new index column so as to be able to reorganise my data after the spliting and normalisation
del df[df.columns[0]]
df
```

J'applique la standardisation sur mon jeu de données car les colones ont des significations/unités differentes

Tout d'abord, je vais extraire les valeurs des colonnes d'attributs des vagues 1-5, 10-21 et des vagues 6-9 pour les normaliser entre 0 et 10, car elles ont été évaluées selon des échelles différentes, d'après ce que j'ai pu lire dans le fichier de description des colonnes.

```python
atribute_df=df[['new','wave','pf_o_att','pf_o_sin','pf_o_int','pf_o_fun','pf_o_amb','pf_o_sha']]
# atribute_df['new'] = atribute_df.index # I add a new index column so as to be able to reorganise my data after the spliting and normalisation
atribute_df
```

Je commence avec ceux des vagues 6-9

```python
wave6_9=atribute_df[atribute_df['wave'].isin([6,7,8,9])]
wave6_9
```

```python
# copy the data
z_scaled_wave6_9 = wave6_9[['pf_o_att','pf_o_sin','pf_o_int','pf_o_fun','pf_o_amb','pf_o_sha']].copy()
ind_wave6_9=wave6_9[['new']]
ind_wave6_9
```

```python
# apply normalization techniques
def standardise(dataset):
  for column in dataset.columns:
      dataset[column] = (dataset[column] -
                            dataset[column].mean()) / dataset[column].std()  
  return dataset 
```

```python
df6_9=standardise(z_scaled_wave6_9)
df6_9
```

```python
#function to normalise the dataset
def normalise(dataset):
  x = dataset.values #returns a numpy array
  min_max_scaler = preprocessing.MinMaxScaler()
  x_scaled = min_max_scaler.fit_transform(x)
  output_df = pd.DataFrame(x_scaled)
  return output_df
```

```python
df6_9_n=normalise(z_scaled_wave6_9) 
df6_9_n.insert(0, "ind", ind_wave6_9) #reinsert the index column
df6_9_n
```

```python
df6_9.describe() #Stahdardised dataset description
```

```python
df6_9_n.describe() #Stahdardised dataset description
```

J'ai décidé de garder la version normalisée plutôt que la version standardisée parce que dans les versions normalisées, les valeurs sont toutes comprises entre 0 et 1 indépendamment des sélections de vagues, alors que dans le format standardisé, les valeurs des différents ensembles de données de vagues varient différemment, c'est-à-dire qu'elles ne sont toujours pas sur la même échelle.

Je répète la même chose avec les colonnes du reste des vagues

```python
wave1_21= atribute_df[atribute_df.wave.isin([6,7,8,9]) == False]
# df.drop(df.loc[df['line_race']==0].index, inplace=True)
wave1_21
```

```python
z_scaled_wave1_21 = wave1_21[['pf_o_att','pf_o_sin','pf_o_int','pf_o_fun','pf_o_amb','pf_o_sha']].copy()
ind_wave1_21=wave1_21[['new']] # indexes of the wave 1-5 ,10-21

z_scaled_wave1_21
ind_wave1_21
```

```python
df1_21=normalise(z_scaled_wave1_21)
df1_21.insert(0, "ind", ind_wave1_21) #reinsert the index column
df1_21
```

```python
df1_21.describe()
```

```python
df_all_waves = pd.concat([df1_21, df6_9_n])
df_all_waves
```

```python
#rearrange the rows by index
pref_df=df_all_waves.sort_values(by='ind',ascending=True)
pref_df.columns = ['new','pf_o_att', 'pf_o_sin', 'pf_o_int', 'pf_o_fun','pf_o_amb','pf_o_sha'] #rename all the columns back to their original name
pref_df
```

```python
pref_df=pref_df.set_index('new') #set column ind as index
pref_df
```

```python
pref_df.describe()
```

Maintenant que nous avons normalisé toutes les évaluations des attributs des différentes vagues, nous pouvons standardiser l'ensemble des colonnes dans l'ensemble du jeu de données. 

```python
to_drop=['pf_o_att','pf_o_sin','pf_o_int','pf_o_fun','pf_o_amb','pf_o_sha']
df1=df.drop(axis=1, columns=to_drop)
df1
```

Je supprime également les données catégorielles, qui seront rajoutées après la normalisation. Aussi, j'enleve les colonnes "iid", "pid", et "gender". Je ne crois pas qu'on aura besoin de les standardiser

```python
cat_cols=['undergra','from','zipcode','iid','pid','gender','dec_o','match']
cat_data=df[cat_cols] 
num_data=df1.drop(axis=1, columns=cat_cols)
num_data = pd.merge(num_data, pref_df, how='inner', on = 'new') #inner joint 
```

```python
num_data=num_data.drop(axis=1, columns='new') # we can now delete the column new
num_data
```

```python
std_df=standardise(num_data)
std_df
```

Je pense que nous n'aurons plus besoin de la colonne "wave", car nous avons déjà normalisé les valeurs des vagues pour qu'elles soient sur la même échelle. 

```python
std_df=std_df.drop(axis=1, columns='wave') # we can now delete the column new
std_df
```

Maintenant, nous pouvons rassembler toutes les pièces pour l'ensemble de données de notre système de recommandation.

```python
final_dataset=pd.concat([cat_data, std_df], axis = 1)
final_dataset.head()
```

Pour regénérer les données pour que chaque ligne corresponde aux action d’un utilisateur (et pas celles du partenaire). Je vai créer un nouveau jeu de données avec des colonnes qui pourraient être remplies par l'utilisateur notament les colonnes "imprace", "imprelig", et à partir des colonnes "match" et "dec_o" je vais créer une colonne "dec_u" correspondant a la décision de de l'utilisateur et pas de son partenair

```python
user_df=final_dataset[['imprace','imprelig','dec_o','match']].copy()
user_df
```

```python
user_df['dec_u']=(abs((user_df["match"] + user_df["dec_o"])-1))
user_df
```

La formule mathématique (somme des colonnes)-1 fonctionne bien pour définir la décision de l'utilisateur. Le seul problème est que lorsque nous avons 0 à la fois à "dec_o" et "match", la décision de l'utilisateur peut être soit 1 soit 0, il est donc difficile de prendre cette décision. Comme le jeu de données est destiné à un système de recommandation sur un site de rencontres, je pense qu'il est préférable de recommander des correspondances potentielles à l'utilisateur et non aux utilisateurs qu'il pourrait aimer, d'où l'utilisation de la colonne "match". Mais comme l'exercice l'exige, je pense qu'il sera bon d'avoir la colonne "dec_u" même si elle ne sera pas très précise.
